```
# imports
import numpy as np
import scipy as sp
import pandas as pd

from plotly.subplots import make_subplots
from plotly import graph_objects as go
import plotly.express as px
import plotly.figure_factory as ff

import matplotlib.pyplot as plt

import plotly.io as pio

pio.templates.default = "plotly_white"
```

```
x0 = np.random.rand(1000)
```

```
fig = go.Figure(data=[
                      go.Histogram(x=x0, histnorm='probability density')
])

fig.update_layout(
    font=dict(size=10),
    height=400,
    width=600
)

fig.show()
```

```
# exponential distribution
lam = 1
x1 = - np.log(1 - x0) / lam

fig = go.Figure(data=[
                      go.Histogram(x=x1, histnorm='probability density', name="generated"),
                      go.Scatter(x=np.linspace(0, 10, 100), y=[np.exp(-t) for t in np.linspace(0, 10, 100)], name="ideal"),
])

fig.update_layout(
    title="Exponential distribution with lambda={0}".format(lam),
    font=dict(size=10),
    height=400,
    width=600
)

fig.show()
```

```
# Cauchy distribution
x2 = np.tan(np.pi * (x0 - 1/2))

fig = go.Figure(data=[
                      go.Histogram(x=[t for t in x2 if -10 < t < 10], name="generated", histnorm='probability density',),
                      go.Scatter(x=np.linspace(-10, 10, 200), y=[1 / (np.pi * (1 + t ** 2)) for t in np.linspace(-10, 10, 200)], name="ideal"),
])

fig.update_layout(
    title="Cauchy distribution".format(lam),
    font=dict(size=10),
    height=400,
    width=400
)

fig.show()
```
